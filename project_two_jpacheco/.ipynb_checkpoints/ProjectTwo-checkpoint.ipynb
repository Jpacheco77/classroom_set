## Question 2. Produce a histogram of the minimum ages of the ant specimens in this dataset (15 pts) . The name of the X-Axis should be "Age of Fossil (millions of years)". 

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(ant_data['min_ma'], color='#0504aa', alpha=0.7, rwidth=0.85)
plt.xlabel("Age of Fossil (millions of years)", size=15)
plt.ylabel("frequency", size=15)
plt.title("Histogram of ant fossil ages", size=15)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Question 3. Produce a bar plot of how many fossils exist for each subfamily. (10 pts). 

```python
my_plot = ant_data['subfamily'].value_counts().plot(kind='bar')
my_plot.set_xlabel("Subfamily", size=15)
my_plot.set_ylabel("Frequency", size=15)
my_plot.set_title('Ant Survey Subfamily Frequence', fontsize=15)

```
