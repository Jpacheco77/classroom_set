## Question 2. Produce a histogram of the minimum ages of the ant specimens in this dataset (15 pts) . The name of the X-Axis should be "Age of Fossil (millions of years)". 

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(ant_data['min_ma'])
plt.xlabel("Age of Fossil (millions of years)")
plt.ylabel("frequency")
plt.title("Histogram of fossil ages")
plt.grid(True)
plt.show()
```

## Question 3. Produce a bar plot of how many fossils exist for each subfamily. (10 pts). 

```python
my_plot = ant_data['subfamily'].value_counts().plot(kind='bar')
my_plot.set_xlabel("Subfamily")
my_plot.set_ylabel("Frequency")
```
